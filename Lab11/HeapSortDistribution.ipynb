# Heap Sort CLRS

```python
def max_heapify(A, i):
    l = 2*i
    r = 2*i+1
    if l <= (len(A)- 1) and A[l] > A[i]:
        largest = l 
    else: 
        largest = i
    if r <= len(A)- 1 and A[r] > A[largest]:
        largest = r
    if largest != i:
        A[i], A[largest] = A[largest], A[i]
        max_heapify(A, largest)
```

```python
def build_max_heap(A):
    for i in range(int(len(A) / 2), 0, -1):
        max_heapify(A, i)
```

## Heap Sort

```python
def heap_sort(A):
    B = []
    build_max_heap(A)
    while len(A)>1:
        maxim = A[1]
        A[1] = A[-1]
        del A[-1]
        B.insert(0,maxim)
        max_heapify(A,1)
    return B
        
```

```python
print(heap_sort([0,4,6,7,2,5,9]))
print(heap_sort([0,3,7,3,8,13,33,4]))
print(heap_sort([0,4,6,7,2,5.5,5.2]))
```

##Heap Sort time complexity histograms for permutations of  n  different elements

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(1,21,1)

plt.plot(n, 5*n-4,  label = 'Heap sort best time $5n-4$')
plt.plot(n, 3/4*n**2+17/4*n-4, label = 'Heap average time $3/4 n*^2+17/4 n - 4$')
plt.plot(n, 3/2*n**2+7/2*n-4, label = 'Heap average time $3/2 n*^2+7/2 n - 4$')
plt.xlabel('Array size')
plt.ylabel('Steps')
plt.legend(loc = 'upper left')
```

```python
def heap_sort_steps(A):
    steps = 0
    B = []
    build_max_heap(A)
    while len(A)>1:
        maxim = A[1]
        A[1] = A[-1]
        del A[-1]
        B.insert(0,maxim)
        max_heapify(A,1)
        steps=steps+5
    steps=steps+1  
    return steps
```

```python
def max_heapify_steps(arr, a, i, steps):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2
    steps += 3

    if left < a and arr[largest] < arr[left]:
        largest = left
        steps += 1

    if right < a and arr[largest] < arr[right]:
        largest = right
        steps += 1

    steps += 2

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        steps += max_heapify_steps(arr, a, largest, steps) + 1

    steps += 1
    return steps


def heap_sort_steps(arr):
    steps = 0
    n = len(arr)

    for i in range(n // 2 - 1, -1, -1):
        steps += max_heapify_steps(arr, n, i, steps)

    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        steps += max_heapify_steps(arr, i, 0, steps)

    steps += 2
    return steps
```
