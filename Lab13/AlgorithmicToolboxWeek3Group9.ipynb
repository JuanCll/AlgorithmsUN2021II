### 3-1: Money Change


```python
n = int(input())
x = 0
for i in [10, 5, 1]:
    if n>=i:
        q = n//i
        x += q
        n = n%i
        if n==0:
            print(x)
            quit()
```

### 3-2: Maximun Value of the Loot (Fractional Knapsack)

```python
n, m = [int(i) for i in input().split()]
lst = []

if m == 0:
    print(0)
    quit()

for h in range(n):
    v, w = [int(i) for i in input().split()]
    if v == 0:
        continue
    lst.append((v, w))

lst.sort(key = lambda x: x[0]/x[1], reverse = True)

t = 0

for v,w in lst:
    if m==0:
        print(t)
        break
    d = min(w, m)
    t += d*v/w
    m -= d

print(t)
```

### 3-3: Car Fueling

```python
d = int(input())
m = int(input())
n = int(input())
lp = list(map(int, input().split()))

lp.append(d)
np = 0
ld = d
cd = 0
for i in range(len(lp) - 1):
    if (ld == 0):
        break
    elif ((lp[i + 1] - lp[i]) > m):
        np = -1
        break
    elif ((cd + m) >= lp[i] and (cd + m) < lp[i + 1]):
        cd = lp[i]
        ld = d - lp[i]
        np +=  1

print(np)
```

### 3-4: Maximum Advertisement Revenue (Maximum Dot Product)

```python
n = int(input())
a = [int(i) for i in input().split()]
b = [int(i) for i in input().split()]
a.sort()
b.sort()

maxa = sum([a[i]*b[i] for i in range(n)])
print(maxa)
```

### 3-5: Collecting Signatures (Covering Segments by Points)

```python
n = int(input())
L = []

for h in range(n):
    a, b = [int(i) for i in input().split()]
    L.append((a,b))

L.sort(key = lambda x: x[1])

i = 0
co = []

while i < n:
    ac = L[i]
    while i < n-1 and ac[1]>=L[i+1][0]:
        i += 1
    co.append(ac[1])
    i += 1
print(len(co))
print(' '.join([str(i) for i in co]))
```

### 3-6: Maximum Number of Prizes (Different Summands)

```python
n = int(input())
if n == 1:
    print(1)
    print(1)
    quit()
m = n
prizes = []
for i in range(1, n):
    if m>2*i:
        prizes.append(i)
        m -= i
    else:
        prizes.append(m)
        break

print(len(prizes))
print(' '.join([str(i) for i in prizes]))
```

### 3-7: Maximum Salary (Largest Number)

```python
n = int(input())
l = list(map(int, input().split()))

sal = []

while l!=[]:
    maxd = 0
    for i in l:
        if int(str(i)+str(maxd))>=int(str(maxd)+str(i)):
            maxd = i
    sal.append(maxd)
    l.remove(maxd)

print(''.join([str(i) for i in sal]))
```
